# Configbox

```python
import yaml

yaml_file_path = "test.yaml"

# 1. You can open the yaml file and convert its data into a dictionary
# Open the YAML file and load its contents
with open(yaml_file_path, 'r') as yaml_file:
    data = yaml.safe_load(yaml_file)


print(data)
print(type(data))
```

```python
# 2. Despite that data has dict data type, you can only access the values with [""]
# 2a. Accessing value with [""]
data["name"]
```

```python
data["Email"]
```

```python
# 2b. Attempting to access the values with dot notation, ., gives error
# data.name # this gives error
```

```python
# 3. To make the dot notation work, like the pandas df, use Box package

from box import ConfigBox

config_data = ConfigBox(data)

config_data.name
```

```python
# 3b. Your trusted [''] also still works 
config_data['name']
```

# ensure_annotations

```python
def get_product(x:int, y:int):
    return x*y
```

```python
get_product(x=2,y=4)
```

```python
# Despite that y is declared int, assigning a string will still work, which ought not be so
get_product(x=2,y="4")
```

```python
# But the strictness of the data type can be ensured with ensure
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int, y:int):
    return x*y
```

```python
get_product(x=2,y=4)
```

```python
# Now, you will get an error
# get_product(x=2,y="4") 
```
